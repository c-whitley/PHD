```python
with open(os.path.join(os.getcwd(), "done.pickle"), "rb") as file:
          
    done = pickle.load(file)

for file_name in glob.iglob(r"H:\FTIR_Data\Conor\ECS\**\*.dms", recursive  = True):
    
    print(file_name)
    
    if file_name.split("\\")[-2] in done: continue

    if "Barney" in file_name or "TMA" in file_name:
        continue

    
    output = eng.agilentFile(file_name, nargout = 6)
    eng.close()
    
    metadata = {}
    
    metadata["wn"] = np.array(output[0], dtype = (np.int))[0]
    image = np.array(output[1])
    metadata["w"] = np.int(output[2])
    metadata["h"] = np.int(output[3])
    metadata["fn"] = output[4]
    metadata["TMA"] = 1
    
    try:
        metadata["name"] = metadata["fn"].split("\\")[-2]
    except:
        pass
    
    metadata["date"] = output[5]
    
    envi.save_image(r"H:\FTIR_Data\numpy_format\ECS\{}.hdr".format(metadata["name"]),
                    image
                   ,force=True
                   ,metadata = metadata)
    
    done.append(metadata["name"])
    
    with open(os.path.join(os.getcwd(), "done.pickle"), "wb") as file:
          
        pickle.dump(done, file)
```

```python
def im_correction(im, i = 2):
    
    im_s = np.split(im, i, 0)
    im_s = np.array([np.split(im, 2, 1) for im in im_s])
    
    #return im_s

    im_std = im_s
    im_mean = im_s
    
    for i in range(i):
        im_std = im_std.std(axis = 0)
        im_mean = im_mean.mean(axis = 0)
        
    return im_std
```

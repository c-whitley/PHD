```python
import os

from scipy.io import loadmat
from skimage.io import imsave

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
os.listdir(os.getcwd())
```

```python
data = loadmat(os.path.join(os.getcwd(), 'Tissue FTIR images.mat'))
```

```python
data.keys()
```

```python
g1 = data["G1data"]
h1 = data["H1data"]
wn = data["Wavenumbers"]
```

```python
pd.DataFrame(wn).T.to_excel(os.path.join(os.getcwd(), "Wavenumbers.xlsx"))
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2)

ax1.imshow(g1.mean(axis = 2))
ax2.imshow(h1.mean(axis = 2))
```

```python
imsave(os.path.join(os.getcwd(), "G1.jpeg"), g1.mean(axis = 2))
imsave(os.path.join(os.getcwd(), "H1.jpeg"), h1.mean(axis = 2))
```

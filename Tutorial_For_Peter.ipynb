```python
import os
import glob
import pickle
import string
from tqdm.notebook import tqdm

import numpy as np
import seaborn as sns
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib
from matplotlib.patches import Rectangle
import matplotlib.colors as mcolors
```

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.svm import SVC, LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
X, Y = make_classification(n_features=3,n_clusters_per_class=1,n_classes=3, n_informative=3, n_redundant=0)
```

```python
import plotly.graph_objects as go
import numpy as np


x, y, z = [comp.squeeze() for comp in np.split(X, 3, 1)]

fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers', marker=dict(
        size=5,
        color=Y,                # set color to an array/list of desired values
        opacity=1
    ))])

fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
fig.show()
```

# PCA

```python
fig, ax = plt.subplots(figsize=(6,6))

pc = PCA(n_components=2)
new_X = pc.fit_transform(X)

for i, y_g in enumerate(np.unique(Y)):

    ax.scatter(new_X[Y==y_g, 0], new_X[Y==y_g, 1])

ax.set_xlabel(f'Principal component 1: explained_variance: {pc.explained_variance_ratio_[0]:0.2f}')
ax.set_ylabel(f'Principal component 2: explained_variance: {pc.explained_variance_ratio_[1]:0.2f}')
```

# LDA

```python
fig, ax = plt.subplots(figsize=(6,6))

lda = LinearDiscriminantAnalysis(n_components=2)
new_X = lda.fit_transform(X, Y)

for i, y_g in enumerate(np.unique(Y)):

    ax.scatter(new_X[Y==y_g, 0], new_X[Y==y_g, 1])
```

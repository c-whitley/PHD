```python
files = {file_name.split("HEADR")[0][:-1]: SNOM_Import.SNOM_File(file_name) for file_name in glob.glob(file_place)}

output = []

for file_name, file_object in files.items():
    
    results = dict()
    
    for line in file_object.HEADR_RAW.split("\n"):
        
        results["File Loc"] = file_name

        if "=" in line:

            name, value = line.split("=")[0], line.split("=")[1]

            results[name.capitalize().strip()] = value
            
            
        else:
            
                if "Pulse Width" in line:

                    _, value = line.split("=")[0], line.split("Pulse Width")[1]

                    results["Pulse Width"] = value
            
            
    output.append(results)
```
